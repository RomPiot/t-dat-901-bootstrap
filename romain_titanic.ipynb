```python
# data_range = pd.cut(data['age'], bins=[0, 9, 19, 29, 39, 49, 59, 69, 79, 89, 99], include_lowest=True)
# data_range = data['age'].map(lambda x:x+10)
# data_range.describe()

# data.loc[data['age'] <= 9, 'age'] = 0
# data.loc[(data['age'] > 9) & (data['age'] <= 19), 'age'] = 1
# data.loc[(data['age'] > 19) & (data['age'] <= 29), 'age'] = 2
# data.loc[(data['age'] > 29) & (data['age'] <= 39), 'age'] = 3
# data.loc[(data['age'] > 39) & (data['age'] <= 49), 'age'] = 4
# data.loc[(data['age'] > 49) & (data['age'] <= 59), 'age'] = 5
# data.loc[(data['age'] > 59) & (data['age'] <= 69), 'age'] = 6
# data.loc[(data['age'] > 69) & (data['age'] <= 79), 'age'] = 7
# data.loc[(data['age'] > 79) & (data['age'] <= 89), 'age'] = 8
# data.loc[(data['age'] > 89) & (data['age'] <= 99), 'age'] = 9

def categorize_age(age):
    if (age <= 9.0): return '[0-9]'
    if (age > 9.0) & (age <= 19.0): return '[10-19]'
    if (age > 19.0) & (age <= 29.0): return '[20-29]'
    if (age > 29.0) & (age <= 39.0): return '[30-39]'
    if (age > 39.0) & (age <= 49.0): return '[40-49]'
    if (age > 49.0) & (age <= 59.0): return '[50-59]'
    if (age > 59.0) & (age <= 69.0): return '[60-69]'
    if (age > 69.0) & (age <= 79.0): return '[70-79]'
    if (age > 79.0) : return '[80+]'

data_range = data['age'].map(categorize_age)
data.groupby(data_range).mean()
```
